# Linear Quadratic dynamic programming

Consider the following Linear-Quadratic problem:
\begin{equation*}
\max_{u_t}  - \sum_{t=0}^\infty \beta^t x^T_t R x_t + u^T_t Q u_t + 2 u_t^T N x_t\\
\quad s.t. \quad x_{t+1} = A x_t + B u_t + C \epsilon_t
\end{equation*}
where $\epsilon_t$ is iid with mean zero.

In recursive form, the functional equation reads:
\begin{equation*}
v(x) = \max_{u}  -  x^T R x - u^T Q u - 2 u^T N x - \beta v(A x + B u + C \epsilon ) \\
\end{equation*}

This problem can be solved by the method of undetermined coefficients:
- guess that $v$ is afine quadratic with a constant: $v(x) = x^T P x + d$
- subsitute our guess in the functional equation
- take FOC w.r.t $u$ to get the $P$-greedy (linear) policy function 
- substitute the linear policy function in the functional equation to get rid of the max; in the LQ case, the resulting equation is an Algebraic Riccati equation

For the finite horizon case:
- first determine the final conditions $x^T R_f x$ (either given or solve the static problem at the end)
- compute $F_{T-1}$ taking $R_f$ as the future value function $R_f := P_T$.
- compute $P_{T-1}$ by plugging $R_f$ as $P_T$ in the Algebraic Riccati equation to determine the value function $P_{T-1}$
- continue by induction.

For the infinite horizon case:
- solve the Riccati equation for $P$ by making an initial guess for $P^0$ and iterating until convergence.

Note: the infinite horizon solution is the limit of the finite horizon solution. In other words, the first $T$ elements of the sequence $P^n$ (the limit of which is the solution for the infinite horizon case) is the same as the sequence $R_f, P_{T-1}, \ldots, P_0$ for the finite horizon case if the initial guess (which shouldn't matter) is $P^0 = R_f$. For the finite horizon case, for every element of the sequence of value functions, the corresponding policy function is computed. For the infinite horizon case, only the policy function of the fixed point of the Riccati equation, namely $P = lim_{n \to \infty} P^n$ is computed.

Remark:
If the transition function for the state variable is afine in $x_t, u_t$, it can be linearized at the cost of an extra state variable, that is, stacking a constant at the top of $x_t$.

Result: (Certainty Equivalence)
In the LQ problem, the optimal policy function is independent of $C \epsilon$ although the value fucntion depends on $C \epsilon$. We can therefore solve the LQ problem without shocks to get the policy function and just add the shock when simulating $x_t$.

The solution for the infinite horizon case is:
- policy function $u = -F x$ where $F = (Q +\beta B^TPB)^{-1} \beta (B^T P A + N)$
- value function $v(x) = x^T P x + d$ where $d = \frac{\beta}{1-\beta} trace(C^T PC)$ and 
$P$ solves the Algebraic Riccati Equation:
\begin{equation*}
P = R−(\beta B^T P A + N)^T(Q+ \beta B^T P B)^{−1} (\beta B^T P A + N)+ \beta A^T P A
\end{equation*}
- the state evolves according to:
$x_{t+1} = A x_t - B F x_t + C \epsilon_t$

```python
import matplotlib.pyplot as plt
%matplotlib inline
from __future__ import division
import numpy as np
from numpy.linalg import inv
from numpy import dot

'''
Solution to LQ problem
'''
def LQ_Fpolicy(beta, P, A, B, Q, N):
    return dot(inv(Q + beta * dot(B.T, dot(P,B))), (beta * dot(B.T, dot(P,A)) + N))

def LQpolicy(F, x):
    return - dot(F,x)

def LQvalue(x, P, beta, C):
    d = (beta /(1 - beta)) * np.trace(dot(C.T, dot(P,C)))
    return dot(x.T, dot(P,x)) + d

def next_state(x, A, B, C, epsilon, F):
    return dot((A - dot(B,F)),x) + C * epsilon

def update_Riccati(P, beta, A, B, R, Q, N):
    Z = (beta * dot(B.T, dot(P,A))) + N
    W = inv(Q + beta * dot(B.T, dot(P,B)))
    return R - dot(Z.T, dot(W, Z)) + beta * dot(A.T, dot(P,A))

def Riccati_solution(init_P, beta, A, B, R, Q, N):
    dist = 1
    while dist > 10e-10:
        next_P = update_Riccati(init_P, beta, A, B, R, Q, N)
        dist = np.linalg.norm(init_P - next_P)
        init_P = next_P
    return init_P
```

Below we compute an inifinite horizon consumption-savings LQ dynamic programming problem.

Consider the following economic problem:
\begin{equation*}
v(a, y) = \ \ min (c - \bar{c})^2 + \beta \mathbb{E} v(a', y')\\
s.t. \quad a' = (1+r) a - c + y\\
y = \mu + \sigma \epsilon, \quad \epsilon \sim N(0, 1)
\end{equation*}

It can be cast as a LQ programming problem with:
- $u = c-\bar{c}$, 
- $R = 0$, $Q = 1$, $N=0$,
- $x^T = [a, 1]$
- $A = [[1+r, -\bar{c}+\mu],[0,1]]$
- $B^T = [-1, 0]$
- $C^T = [\sigma, 0]$.

The consumer problem does not feature a no-ponzi scheme and therefore nothing prevents the consumer from accumulating debt. The value function still converges because the consumer is only trying to minimize a loss function and the loss function is not strictly decreasing in consumption.

```python
'''
Environment
'''
# Parameters
q = 10e6 #this is like a transversality condition: don't save too much
r = 0.05
beta = 1/(1+r)
Cbar = 2
mu = 1
sigma = 0.25

# Elements of the LQ programming problem
n = 2 #number of state variables
k = 1 #number of control variables
R = np.zeros((n,n))
Rf = np.array([[q, 0],[0, 0]]) #the final cost of the state variable
Q = np.ones(k)
N = np.zeros((k,n))
A = np.array([[1+r, -Cbar + mu],[0, 1]])
B = np.array([[-1],[0]])
C = np.array([[sigma],[0]])

'''
Main loop for computations of solution
'''
init_P = np.ones((n,n))

P = Riccati_solution(init_P, beta, A, B, R, Q, N)
F = LQ_Fpolicy(beta, P, A, B, Q, N)

'''
Simulation
'''
T = 45
a_0 = 0 # initial assets
Savings = np.empty(T) # records savings
Consumption = np.empty(T) # records consumption
Wealth = np.empty(T) # records wealth

epsilon = np.random.normal(0,1,T) #np.zeros(T) 
x = np.array([[a_0], [1]])

for i in range(T):
    u = LQpolicy(F, x)
    c = u + Cbar
    x = next_state(x, A, B, C, epsilon[i], F)
    Savings[i] = x[0]
    Consumption[i] = c
    if i > 0:
        Wealth[i] = mu + sigma * epsilon[i] + (1+r) * Savings[i-1]
    else:
        Wealth[i] = mu + sigma * epsilon[i]
        
#print Consumption, 'Consumption'
#print Assets, 'Assets'

fig, ax = plt.subplots()
ax.plot(range(T), Consumption, linewidth=1.3, label='Consumption', alpha=0.7)
#ax.plot(range(T), Savings, linewidth=1.3, label='Savings', alpha=0.7)
ax.plot(range(T), Wealth, linewidth=1.3, label='Wealth', alpha=0.7)
ax.legend(loc='lower left')
ax.set_title('Consumption-Saving for an LQ problem')
#ax.set_ylim(-1, 1.5)
#ax.set_xlim(np.min(k), np.max(k))
plt.show()
```

Below we compute a finite horizon consumption-savings LQ dynamic programming problem.

Unlike the infinite horizon, the solution is computed by backward induction.

Again,
- $u = c-\bar{c}$, 
- $R = 0$, $Q = 1$, $N=0$,
- $R^f$ = [[q, 0],[0,0]]
- $x^T = [a, 1]$
- $A = [[1+r, -\bar{c}+\mu],[0,1]]$
- $B^T = [-1, 0]$
- $C^T = [\sigma, 0]$.

The solution is:
- policy function $u = -F_t x$ where $F_t = (Q +\beta B^T P_{t+1} B)^{-1} \beta (B^T P_{t+1} A + N)$
- value function $v_t(x) = x^T P_t x + d_t$ where $d_t = \beta (d_{t+1} + trace(C^T PC))$ and 
$P_t$ is given by:
\begin{equation*}
P_t = R−(\beta B^T P_{t+1} A + N)^T(Q+ \beta B^T P_{t+1} B)^{−1} (\beta B^T P_{t+1} A + N)+ \beta A^T P_{t+1} A
\end{equation*}
- the state evolves according to:
$x_{t+1} = A x_t - B F_t x_t + C \epsilon_t$

and $P_T = R_f$, $d_T = 0$.

```python
'''
Environment
'''
# Parameters
q = 10e6 #this is like a transversality condition: don't save too much
r = 0.05
beta = 1/(1+r)
Cbar = 2
mu = 1
sigma = 0.25

# Elements of the LQ programming problem
n = 2 #number of state variables
k = 1 #number of control variables
R = np.zeros((n,n))
Rf = np.array([[q, 0],[0, 0]]) #the final cost of the state variable
Q = np.ones(k)
N = np.zeros((k,n))
A = np.array([[1+r, -Cbar + mu],[0, 1]])
B = np.array([[-1],[0]])
C = np.array([[sigma],[0]])

'''
Main loop for computations of solution
'''
T = 45
a_0 = 0 # initial assets
Savings = np.empty(T) # records savings
Consumption = np.empty(T) # records consumption
Wealth = np.empty(T) # records wealth
Income = np.empty(T)
CumulIncomeShock = np.empty(T)
Pfinite = []
Ffinite = []
dfinite = []
Pfinite.append(Rf)
#Ffinite.append(LQ_Fpolicy(beta, Rf, A, B, Q, N))
dfinite.append(0)

x = np.array([[a_0], [1]])
Z = 0 #initial cumulated income shocks

for i in range(T):
    F = LQ_Fpolicy(beta, Pfinite[i], A, B, Q, N)
    Ffinite.append(F)
    P = update_Riccati(Pfinite[i], beta, A, B, R, Q, N)
    Pfinite.append(P)
    d = beta * (dfinite[i] + np.trace(dot(C.T, dot(P,C))))
    dfinite.append(d)

for i in range(T):
    u = LQpolicy(Ffinite[T-1-i], x)
    c = u + Cbar
    epsilon = np.random.normal(0,1,1)
    z = sigma * epsilon
    Z = Z + z
    CumulIncomeShock[i] = Z
    y = mu + sigma * epsilon
    x = next_state(x, A, B, C, epsilon, Ffinite[T-1-i])
    Savings[i] = x[0]
    Consumption[i] = c
    Income[i] = y
    if i > 0:
        Wealth[i] = mu + sigma * epsilon + (1+r) * Savings[i-1]
    else:
        Wealth[i] = mu + sigma * epsilon
   
fig, ax = plt.subplots()
fig, ax1 = plt.subplots()
ax.plot(range(T), Consumption, 'k-', linewidth=1.3, label='Consumption', alpha=0.7)
ax.plot(range(T), Income, 'g-', linewidth=1.3, label='Non financial Income', alpha=0.7)
ax1.plot(range(T), Savings, 'b-', linewidth=1.3, label='Assets', alpha=0.7)
#ax1.plot(range(T), Wealth, 'y-', linewidth=1.3, label='Assets', alpha=0.7)
#ax1.plot(range(T), Income, 'g-', linewidth=1.3, label='Non financial Income', alpha=0.7)
ax1.plot(range(T), CumulIncomeShock, 'r-', linewidth=1.3, label='cumulative unanticipated income', alpha=0.7)
ax.legend(loc='lower left')
ax1.legend(loc='upper left')
ax.set_title('Consumption-Saving for a finite horizon LQ problem')
ax.set_ylim(0.4, 1.8)
ax1.set_ylim(-3, 3)
#ax.set_xlim(np.min(k), np.max(k))
plt.show()

'''
print Pfinite, 'P'
print Ffinite, 'F'
print dfinite, 'd'
print Consumption, 'C'
print Savings, 'a'
'''
```

Below we solve Exercise 3 of QuantEcon LQ Dynamic Programming http://quant-econ.net/py/lqcontrol.html#solution-lqc-ex1-fig-py

Model:
\begin{equation*}
p_t = a_0 - a_1 q_t + d_t
\end{equation*}
with demand shocks: $d_t = \rho d_{t-1} + \sigma w_{t+1}$.

Profits are 
\begin{align*}
\pi_t =& (p_t - c) q_t - \gamma (q_{t+1}- q_t)^2\\
=& (a_0 - c - a_1 q_t + d_t) q_t - \gamma (q_{t+1}- q_t)^2\\
=& (a_0 - c) q_t - a_1 q_t^2 + d_t q_t - \gamma q_{t+1}^2 - \gamma q_t^2 + 2 \gamma q_{t+1} q_t\\
=& (a_0 - c) q_t - (a_1 + \gamma) q_t^2 + d_t q_t - \gamma q_{t+1}^2  + 2 \gamma q_{t+1} q_t
\end{align*}

The firm solves:
\begin{equation*}
\max_{(p_t)_{t=0}^\infty} \quad \mathbb{E} \sum^\infty_{t=0} \beta^t \pi_t
\end{equation*}

The strength of the adjustment cost is measured by $\gamma$.

If $\gamma = 0$, the firm faces a static decision problem every period and the solution is:
\begin{equation*}
\bar{q}_t = \frac{a_0 - c + d_t}{2 a_1}.
\end{equation*}

Recursive formulation attempt 1:

\begin{equation*}
v(q, d, 1) = \max_{q'} \ \big((a_0 - a_1 q + d) - c\big) q - \gamma (q'- q)^2 + \beta \mathbb{E} v(q', d', 1)
\end{equation*}
Initial state is $d_0 = 0$ and $q_0 = 0$ and $d' = \rho d + \sigma \epsilon$ where $\epsilon \sim N(0,1)$.

Note that in this solution, the firm sets its price before observing next period's shock. The adjustment cost together with the persistence causes the firm to adjust with a lag so its production path exhibits variations of larger amplitude than without adjustment costs.

```python
'''
Environment
'''
# Parameters
gamma = 0
a_0 = 5
a_1 = 0.5
rho = 0.9
sigma = 0.15
beta = 0.95
c = 2 

def qbar(a_0, a_1, c, d):
    '''
    Solution without adjustment cost:
    '''
    return (a_0 - c + d)/ (2 * a_1)

# Elements of the LQ programming problem
n = 3 #number of state variables
k = 1 #number of control variables
#Rf = 0 #infinite horizon problem
R = np.array([[gamma + a_1, -1, 0], [0, 0, 0], [c - a_0, 0, 0]])
Q = np.array([gamma])
N = np.array([-gamma, 0, 0])
A = np.array([[0, 0, 0], [0, rho, 0], [0, 0, 1]])
B = np.array([[1], [0], [0]])
C = np.array([[0], [sigma], [0]])

'''
Main loop for computations of solution
'''
init_P = np.ones((n,n))

P = Riccati_solution(init_P, beta, A, B, R, Q, N)
F = LQ_Fpolicy(beta, P, A, B, Q, N)

'''
Simulation
'''
T = 50
Production = np.empty(T) # records production
Productionbar = np.empty(T) # records production without 0 adjustment cost
Price = np.empty(T) # records prices

epsilon = np.random.normal(0,1,T+1) 
x = np.array([[0], [sigma * epsilon[0]], [1]])
dshock = 0

for i in range(T):
    dshock = rho * dshock + sigma * epsilon[i]
    qb = qbar(a_0, a_1, c, dshock)
    q = LQpolicy(F, x)
    Production[i] = q
    Productionbar[i] = qb
    x = next_state(x, A, B, C, epsilon[i+1], F)

print F

fig, ax = plt.subplots()
ax.plot(range(T), Production, linewidth=1.3, label='Production', alpha=0.7)
ax.plot(range(T), Productionbar, linewidth=1.3, label='Production $0$ adj cost', alpha=0.7)
ax.legend(loc='lower right')
ax.set_title('Monopolist LQ problem')
#ax.set_ylim(-1, 1.5)
#ax.set_xlim(np.min(k), np.max(k))
plt.show()
```

There is a problem with the above solution: the third entry of $F$ should not be a zero; this zero makes the solution computed here be a shifted version of the actual solution.

Recursive formulation attempt 2:

\begin{equation*}
v(q_{-}, d, 1) = \max_{q} \ \big((a_0 + a_1 q + d) - c\big) q - \gamma (q - q_{-})^2 + \beta \mathbb{E} v(q, d', 1)
\end{equation*}
Initial state is $d_0 = 0$ and $q_{-1} = 0$ and $d' = \rho d + \sigma \epsilon$ where $\epsilon \sim N(0,1)$.

Recall that 
\begin{align*}
\big((a_0 + a_1 q + d) - c\big) q - \gamma (q - q_{-})^2 &= (a_0 - c) q + (a_1 - \gamma) q^2 + d q - \gamma q_{-}^{2}  + 2 \gamma q_{-} q
\end{align*}

Nothing changes in terms of the law of motion compared to the first attempt.
