# Uncertainty Traps

Real Business Cycle models do not generate long lasting recessions as we observe in the data. Fajgelbaum, Schaal and Taschereau-Dumouchel (2015) provide a mechanism which generate persistence in recessions. Their model features:
- changes along the extensive margine: there is a pool of entrepreneurs; some are active, some are inactive;
- inferring from the actions of others: entrepreneurs infer the economic conditions from observing the outcomes of active entrepreneurs (information externality);

Combining the above two features, during a recession, less entrepreneurs are active (movement along the extensive margin), which makes the signal about economic conditions to inactive entrepreneurs more noisy than it would be if there were no changes along the extensive margin. Due to higher uncertainty faced by inactive entrepreneurs, fewer become active which makes the recession persist.

The model:
- Evolution of fundamentals, which are not observable by any agent:
\begin{equation*}
\theta_t = \rho \theta_{t-1} + \sigma_\theta w_t \quad \quad w_t \sim_{iid} N(0,1)
\end{equation*}
$\sigma_\theta > 0$ and $0 < \rho < 1$.


- Signals about fundamentals, the precision of which depends on the mass of active entrepreneurs:
Each active entrepreneur $m$, produces $x_{mt}$ which is a signal of $\theta$ for inactive entrepreneurs:
\begin{equation*}
x_{mt} = \theta_t + \epsilon_{mt} \quad \quad \epsilon_{mt} \sim_{iid} N(0, 1/\gamma_x)
\end{equation*}


- Updating of beliefs about $\theta_{t+1}$:
All agents start with the same initial beliefs and have identical beliefs throughout time since they observe the same signals. Heterogeneity in participation will come from idiosyncratic cost shocks.
The initial distribution of beliefs is normally distributed and the $M$ signals generated by active entrepreneurs are also normally distributed. We can therefore use the Kalman filter to determine the predictive distribution of $x_{t+1}$ (that is, of $\theta_{t+1}$).
Prior on $\theta_0$ is $N(\mu, 1/\gamma)$.
Given a prior on $\theta_t$ that is say $N(\mu, 1/\gamma)$, the predictive (Kalman) distribution of $\theta_{t+1}$ is $N(\mu', 1/\gamma')$ as follows:
\begin{align*}
\mu' &= \rho \frac{\gamma \mu + \gamma_x M \frac{1}{M}\sum^M_{m=1} x_m}{\gamma + \gamma_x M}\\
\gamma' &= \bigg(\frac{\rho^2}{\gamma + M \gamma_x} + \sigma_{\theta}^2 \bigg)^{-1}
\end{align*}
$\mu'$ is a an average of the prior and the new info, weighted by their informativeness.


- Extensive margin:
Each entrepreneurs $m$ has the same opportunity cost $c$ but an idiosyncratic fixed cost $F_{mt}$ drawn every period. Note that $F_m$ is the only source of heterogeneity across entrepreneurs and therefore drives all the movement along the extensive margin. An entrepreneur is active in period $t$ if
\begin{equation*}
\mathbb{E} [u( x_{mt} - F_{mt}) ] > c
\end{equation*}
where $F_{mt}$ is known at the beginning of period $t$ and the expectation is taken w.r.t. the Kalman predicitive distribution for $x_{mt}$ given information up until $t-1$.
With CARA utility: $(1/a) \cdot (1 - \exp(-a \cdot x))$, we get:
\begin{align*}
c <& \mathbb{E} [u( x_{mt} - F_{mt}) ] \\
=& \frac{1}{a} \bigg(1 - \mathbb{E} [\exp(- a (x_{mt} - F_{mt})) ]\bigg) \\
=& \frac{1}{a} \bigg(1 - \exp\bigg(- a \mu + a F_{mt} + a^2\frac{1/\gamma + 1/\gamma_x}{2}\bigg) \bigg) - c + c \\
=& \psi(F_m, \mu, \gamma, \gamma_x, a, c) + c .
\end{align*}

Since $x_{mt} = \theta + \epsilon_{mt}$, $x_{mt}$ is distributed $N\left(\mu, \frac{1}{\gamma} + \frac{1}{\gamma_x}\right)$. It follows that 
\begin{equation*}
- a (x_{mt} - F_{mt}) \sim N\left(- a (\mu - F_m), a^2 \left( \frac{1}{\gamma} + \frac{1}{\gamma_x} \right) \right).
\end{equation*}
Recall that if $z \sim N(\bar{z}, \sigma^2_z)$, then $e^{z}$ is Lognormally distributed with mean $e^{\bar{z} + \frac{1}{2} \sigma_z^2}$

```python
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
import numpy as np
from numpy.linalg import inv

def AR_one(T, x_0, alpha, sigma, distribution):
    '''
    Returns an np.array of length T for the AR1 model: x_t = alpha * x_{t-1} + sigma * w_t
    x_0: initial value
    sigma: stdandard deviation of the shock
    alpha: persistence parameter
    distribution: is the distribution of the shock
    '''
    x = x_0
    time_series = np.empty(T)
    for i in range(T):
        time_series[i] = x
        epsilon = np.random.normal(0,1)
        xone = alpha * x + sigma * epsilon
        x=xone
    return time_series

# == Returns the mean and precision of Kalman's predictive distribution ==
def KalmanPredictDist_Dim1(M, AverageSignal, rho, sigma_theta, mu, gamma, gamma_x):
    '''
    Returns the posterior mean and precision for the (Kalman) predictive distribution
    M: number of signals
    AverageSignal: Average of the M signals received
    rho: persistence of the state
    sigma_theta: variance of the shock to the state
    gamma: precision of prior belief about the state
    gamma_x: precision of each of the M signals.
    '''
    mu_posterior = (gamma * mu + gamma_x * M * AverageSignal) / (gamma + gamma_x * M)
    mu_prime = rho * mu_posterior
    var = (rho**2 / (gamma + M * gamma_x)) + sigma_theta**2
    gamma_prime = 1/var
    return mu_prime, gamma_prime

# == gamma' as a function of gamma ==
def Gammaprime(M, rho, sigma_theta, gamma, gamma_x):
    '''
    Returns the posterior precision for the (Kalman) predictive distribution
    M: number of signals
    rho: persistence of the state
    sigma_theta: variance of the shock to the state
    gamma: precision of prior belief about the state
    gamma_x: precision of each of the M signals.
    '''
    return 1/(rho**2 / (gamma + M * gamma_x) + sigma_theta**2)


# == psi, the function that determines which entrepreneurs are active ==
def psi(F_m, mu, gamma, gamma_x, a, c):
    '''
    Returns the gain (over opportunity cost) in exepcted (w.r.t Kalman predictive Dist) utility from becoming an active entrepreneur
    F_m: fixed idiosyncratic (only source of heterogeneity) cost of being active this period
    mu, gamma: mean and precision of the prior distribution on production
    gamma_x: precision of the signal
    a: parameter of the CARA utility function
    c: opportunity cost of being active
    '''
    temp = - a * mu + a * F_m + a**2 * (((1/gamma) + (1/gamma_x))/2)
    return (1/a) * (1 - np.exp(temp)) - c

def psiFI(F_m, rho, theta, sigma_theta, a, c):
    '''
    Returns the gain (over opportunity cost) in exepcted (w.r.t Full Info) utility from becoming an active entrepreneur
    F_m: fixed idiosyncratic (only source of heterogeneity) cost of being active this period
    rho: persistence in the AR1 evolution of the state
    theta, sigma_theta: current state and variance to the state next period
    a: parameter of the CARA utility function
    c: opportunity cost of being active
    '''
    temp = - a * rho * theta + a * F_m + a**2 * (sigma_theta**2)
    return (1/a) * (1 - np.exp(temp)) - c
```

Below we plot the posterior precision about $\theta$ as a function of the prior precision for 0 to 6 active signals. The more signals, the higher is the posterior precision.

```python
'''
Parameters
'''
NumFirm = 6 # pool of all (active and inactive) entrepreneurs
a = 1.5 # CARA utility parameter
rho = 0.99 # persistence of the economic fundamentals
sigma_theta = 0.5 # variance of the economic funcdamentals inovation
gamma_x = 0.5 # precision of the signal / inverse of production risk

# == Posterior precision as a function of prior precision ==
fig, ax = plt.subplots(figsize=(10,8))
xgrid = np.linspace(0.01, 3, 200)
ax.plot(xgrid, xgrid, 'k-')
for i in range(NumFirm):
    ax.plot(xgrid, Gammaprime(i, rho, sigma_theta, xgrid, gamma_x), label=r'$M=%d$' % i)
ax.set_title('Posterior precision as a function of prior precision') 
ax.set_xlabel(r'$\gamma$', fontsize=16)
ax.set_ylabel(r"$\gamma '$", fontsize=16)
ax.legend(loc='upper left')
plt.show()
```

Recap of the model:
- Evolution of fundamentals, which are not observable by any agent:
\begin{equation*}
\theta_t = \rho \theta_{t-1} + \sigma_\theta w_t \quad \quad w_t \sim_{iid} N(0,1)
\end{equation*}
$\sigma_\theta > 0$ and $0 < \rho < 1$.

- Participation: participate if draw $F_m \sim N(0,\sigma_F)$ is such that:
\begin{equation*}
\psi(F_m, \mu, \gamma, \gamma_x, a, c) \geq 0
\end{equation*}

- Production/signal of active entrepreneurs 
\begin{equation*}
x_{mt} = \theta_t + \epsilon_{mt} \quad \quad \epsilon_{mt} \sim_{iid} N(0, 1/\gamma_x)
\end{equation*}

- predictive (Kalman) distribution of $\theta_{t+1}$ is $N(\mu', 1/\gamma')$ as follows:
\begin{align*}
\mu' &= \rho \frac{\gamma \mu + \gamma_x M \frac{1}{M}\sum^M_{m=1} x_m}{\gamma + \gamma_x M}\\
\gamma' &= \bigg(\frac{\rho^2}{\gamma + M \gamma_x} + \sigma_{\theta}^2 \bigg)^{-1}
\end{align*}

```python
T = 3000 #number of periods
NumFirm = 200 #number of firms
sigma_F = 1.5
a=1.5 #CARA utility function parameter
gamma_x =0.5   # Production shock precision
rho=0.99       # Correlation coefficient for theta
sigma_theta=0.5# Std dev of theta shock
c=-420         # External opportunity cost
init_mu=0      # Initial value for mu
init_gamma=4   # Initial value for gamma
init_theta=0    # Initial value for theta

# == Generate theta ==
Theta = AR_one(T, init_theta, rho, sigma_theta, np.random.randn) # time series of fundamentals

# == Generate X, a panel of NumFirm firm productions across T periods ==
X = np.empty([NumFirm, T]) # panel of firms productions across time
for m in range(NumFirm):
    for t in range(T):
        epsilon = np.random.normal(0, np.sqrt(1/gamma_x))
        X[m,t] = Theta[t] + epsilon

# == Main Loop ==
NumActive = np.empty(T) # records movement along the extensive margin
TotalProd = np.empty(T)
Beliefs = np.empty(T)
Precision = np.empty(T)
for i in range(T):
    # record beliefs and precision
    Beliefs[i] = init_mu
    Precision[i] = init_gamma
    #determine active firms and average production
    F = sigma_F * np.random.randn(NumFirm) # Draw idiosyncratic fixed costs
    Psi_t = psi(F, init_mu, init_gamma, gamma_x, a, c) # expected gain in utility from becoming active
    x_t = X[:,i]
    Active = (Psi_t > 0).sum() #number of firms active
    NumActive[i] = Active
    TotProd = x_t[Psi_t > 0].sum()
    TotalProd[i] = TotProd
    if Active > 0:
        AvProd = TotProd/Active #average production
    else:
        AvProd = 0
    # update beliefs
    mu_prime, gamma_prime = KalmanPredictDist_Dim1(Active, AvProd, rho, sigma_theta, init_mu, init_gamma, gamma_x)
    init_mu, init_gamma = mu_prime, gamma_prime

fig, ax = plt.subplots(figsize=(10,6))
fig, ax1 = plt.subplots(figsize=(10,2))
fig, ax2 = plt.subplots(figsize=(10,2))
fig, ax3 = plt.subplots(figsize=(10,2))
ax.plot(range(T), Theta, label='Fundamentals')
ax.plot(range(T), Beliefs, label='Beliefs')
ax.set_title('Beliefs and Fundamentals') 
ax.set_xlabel(r'$T$', fontsize=16)
ax.legend(loc='upper left')
ax1.plot(range(T), NumActive, label='Number of Active Entrepreneurs')
ax1.set_title('Number of Active Firms') 
ax2.plot(range(T), Precision, label='Precision of beliefs')
ax2.set_title('Precision') 
ax3.plot(range(T), TotalProd, label='Production')
ax3.set_title('Production') 
plt.show()    
```

Below I compare two economies:
- economy with Kalman filtering (entrepreneurs learn from the outcomes of others). That is the predictive dirstribution is given Kalman filtering.
- economy with conditional expectation (entreprneurs observe $\theta_t$ and decide whether to be active or not based on their expectation of $\theta_{t+1}$ conditional on $\theta_t$. That is, the predictive distribution of $\theta_{t+1}$ is $N(\rho \theta_t, \sigma_\theta^2)$)

```python
T = 3000 #number of periods
NumFirm = 150 #number of firms
sigma_F = 1.5
a=1.5 #CARA utility function parameter
gamma_x =0.5   # Production shock precision
rho=0.99       # Correlation coefficient for theta
sigma_theta=0.5# Std dev of theta shock
c=-420         # External opportunity cost
init_mu=0      # Initial value for mu
init_gamma=4   # Initial value for gamma
init_theta=0    # Initial value for theta

# == Generate theta ==
Theta = AR_one(T, init_theta, rho, sigma_theta, np.random.randn) # time series of fundamentals

# == Generate X, a panel of NumFirm firm productions across T periods ==
X = np.empty([NumFirm, T]) # panel of firms productions across time
for m in range(NumFirm):
    for t in range(T):
        epsilon = np.random.normal(0, np.sqrt(1/gamma_x))
        X[m,t] = Theta[t] + epsilon

# == Main Loop ==
NumActive = np.empty(T) # records movement along the extensive margin
NumActiveFI = np.empty(T) # records movement along the extensive margin
TotalProd = np.empty(T)
TotalProdFI = np.empty(T) # FI: full info
Beliefs = np.empty(T)
Precision = np.empty(T)
for i in range(T):
    # record beliefs and precision
    Beliefs[i] = init_mu
    Precision[i] = init_gamma
    #determine active firms and average production
    F = sigma_F * np.random.randn(NumFirm) # Draw idiosyncratic fixed costs
    Psi_t = psi(F, init_mu, init_gamma, gamma_x, a, c) # expected gain in utility from becoming active
    #Psi_tFI = psi(F, rho * Theta[i], sigma_theta, , a, c) # Full Info expected gain in utility from becoming active
    Psi_tFI = psiFI(F, Theta[i], sigma_theta, a, c)
    x_t = X[:,i]
    Active = (Psi_t > 0).sum() #number of firms active
    ActiveFI = (Psi_tFI > 0).sum() #number of firms active
    NumActive[i] = Active
    NumActiveFI[i] = ActiveFI
    TotProd = x_t[Psi_t > 0].sum()
    TotProdFI = x_t[Psi_tFI > 0].sum()
    TotalProd[i] = TotProd
    TotalProdFI[i] = TotProdFI    
    if Active > 0:
        AvProd = TotProd/Active #average production
    else:
        AvProd = 0
    # update beliefs
    mu_prime, gamma_prime = KalmanPredictDist_Dim1(Active, AvProd, rho, sigma_theta, init_mu, init_gamma, gamma_x)
    init_mu, init_gamma = mu_prime, gamma_prime

fig, ax = plt.subplots(figsize=(10,6))
fig, ax1 = plt.subplots(figsize=(10,2))
fig, ax2 = plt.subplots(figsize=(10,2))
fig, ax3 = plt.subplots(figsize=(10,2))
ax.plot(range(T), Theta, label='Fundamentals')
ax.plot(range(T), Beliefs, label='Beliefs')
ax.set_title('Beliefs and Fundamentals') 
ax.set_xlabel(r'$T$', fontsize=16)
ax.legend(loc='upper left')
ax1.plot(range(T), NumActive, label='Kalman')
ax1.plot(range(T), NumActiveFI, label='Full Info')
ax1.legend(loc='upper left', fontsize=10)
ax1.set_title('Number of Active Firms') 
ax2.plot(range(T), Precision, label='Precision of beliefs')
ax2.set_title('Precision') 
ax3.plot(range(T), TotalProd, label='Kalman')
ax3.plot(range(T), TotalProdFI, label='Full Info')
ax3.legend(loc='upper left', fontsize=10)
ax3.set_title('Production') 
plt.show()    
```
